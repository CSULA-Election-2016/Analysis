#### Average Number of Followers

```python
ave_fol = imp['user_followers_count'].mean()
round(ave_fol, 2)
```

#### Average Number of Friends

```python
ave_fri = imp['user_friends_count'].mean()
round(ave_fri, 2)
```

#### Average Star ratio

```python
# there are some people who doesn't follow anyone, but has many followers
imp.loc[(imp['user_friends_count'] != 0), 'star_ratio'] = imp['user_followers_count'] / imp['user_friends_count']
```

```python
ave_star_ratio = imp['star_ratio'].mean()
ave_star_ratio
#Average Star ratio is ridiculously high - not going to use
```

#### Average Star Ratio 2

```python
ave_star_ratio2 = ave_fol / ave_fri
ave_star_ratio2
```

```python
imp.loc[(imp['user_friends_count'] == 0), 'star_ratio'] = ave_star_ratio2
```

#### Average Number of groups they are in

```python
ave_list = imp['user_listed_count'].mean()
round(ave_list, 2)
```

#### Average Number of Statuses they posted

```python
ave_status = imp['user_statuses_count'].mean()
round(ave_status, 2)
```

#### Average Number of Tweets per Day

```python
def parseDate(row):
    return row['user_statuses_count'] / row['numDays']

    
def getNumDays(row):
    a = parser.parse(row['date'])  
    b = parser.parse(row['user_created_at'])
    return (a-b).days
```

```python
imp['numDays'] = imp.apply(getNumDays, axis=1)
```

```python
imp.head(1)
```

```python
# These are users with no history
# Not reliable
imp.loc[(imp['numDays'] == 0)].head(1)
```

```python
imp2 = imp.loc[(imp['numDays'] != 0)]
```

```python
imp2['tweets_per_day'] = imp2.apply(parseDate, axis=1)
```

```python
ave_tweets_per_day = imp2['tweets_per_day'].mean()
round(ave_tweets_per_day, 2)
```

```python
len(imp2)
```
